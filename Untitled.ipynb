```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
```

```python
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
driver = webdriver.Firefox()
driver.get("https://www.fowaonsite.com/member_directory#!directory/map/ord=lnm")
for _ in range (25):
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    time.sleep(3)
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
results = soup.find_all("a", {'class':'SFcrd'})
```

```python
for member in results:
    print(member)
```

```python
# Create empty list
member_list = []
for member in results:
    # Create empty dict
    data_dict = {}
    # Add title,url and metadata to the disctionary as keys
    name = [name for name in member.find('strong').children]
    try:
        individual_name = name[2].split()
        data_dict['company_name'] = name[0]
    except IndexError:
        individual_name = name[1].split()
        data_dict['company_name'] = "null"
        
    data_dict['first_name'] = individual_name[0]
    data_dict['last_name'] = ' '.join(individual_name[1:])
    data_dict['phone_number'] = member.find_all("div")[1].text
    data_dict['street_address'] = member.find('span', {'itemprop': "streetAddress"}).text
    data_dict['address_locality'] = member.find('span', {'itemprop': "addressLocality"}).text

    # Append data to list
    member_list.append(data_dict)
```
